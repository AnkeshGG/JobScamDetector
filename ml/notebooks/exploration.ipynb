```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/real_or_fake_jobs.csv")
```

```python
# Step 3: Quick look at the data
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())
```

```python

# Step 4: Check column names
print("\nColumns in dataset:")
print(df.columns)
```

```python
# Step 5: Count fraudulent vs real postings
fraud_counts = df['fraudulent'].value_counts()
print("\nFraudulent value counts:")
print(fraud_counts)
```

```python

# Step 6: More descriptive summary
print("\nSummary:")
print(f"Number of REAL postings (0): {fraud_counts.get(0, 0)}")
print(f"Number of FAKE postings (1): {fraud_counts.get(1, 0)}")
```

```python

# Step 7: Optional visualization
import matplotlib.pyplot as plt

fraud_counts.plot(kind='bar', color=['green','red'])
plt.title("Distribution of Real vs Fake Job Postings")
plt.xlabel("Fraudulent (0=Real, 1=Fake)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```
